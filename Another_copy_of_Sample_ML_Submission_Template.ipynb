## ***7. ML Model Implementation***

### ML Model - 1

````python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
from sklearn.linear_model import LogisticRegression
# Fit the Algorithm
model=LogisticRegression()
model.fit(x_train_scaled,y_train)
````

````python
train_pred=model.predict(x_train_scaled)
test_pred=model.predict(x_test_scaled)
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score,f1_score
CM=confusion_matrix(y_train,train_pred)
AA=accuracy_score(y_train,train_pred)
PS=precision_score(y_train,train_pred)
RS=recall_score(y_train,train_pred)
F1=f1_score(y_train,train_pred)
print(f'Confusion Matrix:\n{CM}')
print(f'Accuracy Score:\n{AA}')
print(f'Precision Score:\n{PS}')
print(f'Recall Score:\n{RS}')
print(f'F1 Score:\n{F1}')
````

````python
CM=confusion_matrix(y_test,test_pred)
AA=accuracy_score(y_test,test_pred)
PS=precision_score(y_test,test_pred)
RS=recall_score(y_test,test_pred)
F1=f1_score(y_test,test_pred)
print(f'Confusion Matrix:\n{CM}')
print(f'Accuracy Score:\n{AA}')
print(f'Precision Score:\n{PS}')
print(f'Recall Score:\n{RS}')
print(f'F1 Score:\n{F1}')
````

````python
import matplotlib.pyplot as plt

scores = [AA, PS, RS, F1]
labels = ['Accuracy', 'Precision', 'Recall', 'F1 Score']

plt.figure(figsize=(8,5))

bars = plt.bar(labels, scores, color=['skyblue','lightgreen','salmon','gold'])

plt.ylim(0, 1)
plt.ylabel("Score")
plt.title("Logistic Regression Performance Metrics")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 0.01,
        f'{bar.get_height():.3f}',
        ha='center',
        fontsize=10
    )

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
from sklearn.model_selection import GridSearchCV

log_model = LogisticRegression(random_state=42)

# Hyperparameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l2'],
    'solver': ['lbfgs', 'liblinear'],
    'max_iter': [100, 500, 1000]
}

# GridSearchCV
grid_search = GridSearchCV(
    estimator=log_model,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Fit the Algorithm
grid_search.fit(x_train_scaled, y_train)
print("Best Parameters:", grid_search.best_params_)
print("Best Cross Validation Accuracy:", grid_search.best_score_)

# Best model
best_model = grid_search.best_estimator_

# Predictions
train_pred = best_model.predict(x_train_scaled)
test_pred = best_model.predict(x_test_scaled)
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

print("Training Accuracy :", accuracy_score(y_train, train_pred))
AA=accuracy_score(y_test, test_pred)
print("Testing Accuracy  :", AA)
PS=precision_score(y_test, test_pred)
print("Precision         :", PS)
RS=recall_score(y_test, test_pred)
print("Recall            :", RS)
F1=f1_score(y_test, test_pred)
print("F1 Score          :", F1)

print("\nConfusion Matrix")
print(confusion_matrix(y_test, test_pred))

print("\nClassification Report")
print(classification_report(y_test, test_pred))
````

##### Which hyperparameter optimization technique have you used and why?

GridSearchCV was used to find the best combination of hyperparameters for the Logistic Regression model. It systematically tests different parameter values using 5-fold Cross Validation and selects the combination that gives the highest accuracy. This helps improve model performance and reduces the risk of overfitting.
